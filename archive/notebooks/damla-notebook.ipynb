```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-

"""
SIMAP tender fetcher (single file)
- Iterates over a multilingual keyword list + CPV filters
- Dedupes results across keywords
- Fetches detail JSON per project/publication
- Exports:
    * CSV  -> keeps FULL descriptions
    * PDF  -> truncates long descriptions to avoid ReportLab LayoutError
- PDF is A4 landscape with smaller font and comfy margins

Requires:
    pip install requests reportlab
"""

import csv
import re
import html
import time
import sys
from collections import OrderedDict
from typing import Dict, Iterable, List, Tuple

import requests
from requests.adapters import HTTPAdapter, Retry

from xml.sax.saxutils import escape
from reportlab.platypus import SimpleDocTemplate, Table, TableStyle, Paragraph
from reportlab.lib.pagesizes import A4, landscape
from reportlab.lib.styles import getSampleStyleSheet, ParagraphStyle
from reportlab.lib import colors

# -------------------- CONFIG --------------------

SEARCH_ENDPOINT = "https://prod.simap.ch/api/publications/v2/project/project-search"
DETAIL_ENDPOINT = "https://prod.simap.ch/api/publications/v1/project/{pid}/publication-details/{pub}"

# Multilingual keywords (edit freely)
KEYWORDS: List[str] = [
    "study","studie","étude","etude","studio",
    "analysis","analyse","analisi",
    "economy","economics","wirtschaft","ökonomie","économie","economie","economia",
    "benchmark","benchmarking",
    "wirtschaftsberatung","consulting","conseil","consulenza",
    "72000000","79300000",
    "it-dienste",
    "beratung, software-entwicklung, internet und hilfestellung",
    "markt- und wirtschaftsforschung","umfragen und statistiken",
    "dienstleistung","service","prestation","prestazione",
    "ausschreibung","appel d'offres","appels d'offres","gara","procurement",
    "offenes verfahren","procédure ouverte","procedure ouverte","procedura aperta",
    "bks","bundesamt für statistik","office fédéral de la statistique",
    "office federal de la statistique","ufficio federale di statistica",
    "federal statistical office","ofs","bfs",
    "seco","staatssekretariat für wirtschaft",
    "secrétariat d'état à l'économie","secretariat d'etat a l'economie",
    "segretariato di stato dell'economia",
    "kanton zürich","kanton zuerich","canton de zurich","cantone zurigo",
    "kanton luzern","canton de lucerne","cantone lucerna",
    "kanton","canton","cantone",
    "region","regionen","région","regione",
    "bundesamt für","bundesamt fuer",
    "bundesamt für bauten und logistik","bundesamt fuer bauten und logistik","bbl",
    "index","verbraucherindex","price index","indice des prix","indice dei prezzi",
    "wirtschaftsforschung","economic research","recherche économique","recherche economique","ricerca economica",
    "tunnel",  # original example focus
]

# CPV filter (edit if needed)
CPV_CODES = ["72000000", "79300000"]

# PDF safeguards (avoid LayoutError)
MAX_DESC_CHARS = 700       # PDF ONLY (CSV keeps full text)
PDF_FONT_SIZE  = 8
PDF_LEADING    = 10

# PDF page + layout
PAGE_SIZE     = landscape(A4)
MARGINS       = dict(leftMargin=24, rightMargin=24, topMargin=24, bottomMargin=24)
COLUMN_WIDTHS = [120, 120, 80, 220, 300, 140, 100, 180]  # tune if needed


# -------------------- HELPERS --------------------

def pick(d: Dict, langs: Iterable[str] = ("de","en","fr","it")) -> str:
    """Pick the first non-empty localized value from a dict like {'de': '...', 'en': None, ...}."""
    return next((d.get(k) for k in langs if isinstance(d, dict) and d.get(k)), "") if isinstance(d, dict) else ""

def clean_html(s):
    s = "" if s is None else str(s)
    s = html.unescape(s)
    s = re.sub(r'</p\s*>', '\n', s, flags=re.I)
    s = re.sub(r'<\s*br\s*/?>', '\n', s, flags=re.I)
    s = re.sub(r'<[^>]+>', '', s)  # strip remaining tags
    return s.strip()

def session_with_retries() -> requests.Session:
    s = requests.Session()
    retries = Retry(total=4, backoff_factor=0.4, status_forcelist=(429, 500, 502, 503, 504))
    s.mount("https://", HTTPAdapter(max_retries=retries))
    s.headers.update({"User-Agent": "SIMAPFetcher/1.0 (+https://example.org)"})
    return s

def search_projects(sess: requests.Session, keyword: str, cpv_codes: List[str]) -> List[Dict]:
    params = {"search": keyword, "cpvCodes": cpv_codes}
    r = sess.get(SEARCH_ENDPOINT, params=params, timeout=20)
    r.raise_for_status()
    data = r.json() or {}
    return data.get("projects", []) or []

def fetch_detail(sess: requests.Session, pid: str, pub: str) -> Dict:
    url = DETAIL_ENDPOINT.format(pid=pid, pub=pub)
    r = sess.get(url, timeout=30)
    r.raise_for_status()
    return r.json() or {}

def extract_fields(detail: Dict, list_level_title: str = "") -> Dict[str, str]:
    """Return all fields + both full and truncated descriptions (for CSV/PDF)."""
    pr = detail.get("procurement", {}) or {}

    title = pick(pr.get("projectTitle")) or pick(detail.get("title")) or list_level_title or ""
    desc_full = clean_html(pick(pr.get("orderDescription")))  # CSV uses full
    # PDF truncated version
    if len(desc_full) > MAX_DESC_CHARS:
        desc_pdf = desc_full[:MAX_DESC_CHARS].rstrip() + " … [truncated]"
    else:
        desc_pdf = desc_full

    cpv_label = pick((pr.get("cpvCode") or {}).get("label"), ("en","de","fr","it"))
    process   = pr.get("processType", "") or ""
    pub_date  = detail.get("publicationDate") or ""
    office    = pick(pr.get("procOfficeName"), ("de","en","fr","it")) or pick(detail.get("procOfficeName"))

    return dict(
        title=title,
        desc_full=desc_full,
        desc_pdf=desc_pdf,
        cpv_label=cpv_label or "",
        process=process,
        pub_date=pub_date or "",
        office=office or "",
    )

def write_csv(records: List[Dict[str, str]], path: str) -> None:
    header = ["Project ID", "Publication ID", "Process Type", "Title", "Order Description",
              "CPV Label", "Publication Date", "Procurement Office"]
    with open(path, "w", newline="", encoding="utf-8") as f:
        w = csv.writer(f)
        w.writerow(header)
        for r in records:
            w.writerow([
                r["pid"], r["pub"], r["process"], r["title"],
                r["desc_full"], r["cpv_label"], r["pub_date"], r["office"]
            ])

def write_pdf(records: List[Dict[str, str]], path: str) -> None:
    styles = getSampleStyleSheet()
    tiny = ParagraphStyle(
        "Tiny",
        parent=styles["Normal"],
        fontSize=PDF_FONT_SIZE,
        leading=PDF_LEADING,
        wordWrap="CJK",  # better wrapping for long tokens
    )

    def cell(x: str) -> Paragraph:
        return Paragraph(escape(x or ""), tiny)

    header = ["Project ID", "Publication ID", "Process Type", "Title", "Order Description",
              "CPV Label", "Publication Date", "Procurement Office"]
    table_data = [header]
    for r in records:
        table_data.append([
            cell(r["pid"]), cell(r["pub"]), cell(r["process"]), cell(r["title"]),
            cell(r["desc_pdf"]), cell(r["cpv_label"]), cell(r["pub_date"]), cell(r["office"])
        ])

    doc = SimpleDocTemplate(
        path, pagesize=PAGE_SIZE,
        **MARGINS, title="SIMAP Tenders"
    )

    tbl = Table(table_data, repeatRows=1, colWidths=COLUMN_WIDTHS)
    tbl.setStyle(TableStyle([
        ("BACKGROUND", (0,0), (-1,0), colors.black),
        ("TEXTCOLOR", (0,0), (-1,0), colors.whitesmoke),
        ("FONTNAME", (0,0), (-1,0), "Helvetica-Bold"),
        ("FONTSIZE", (0,0), (-1,0), 9),
        ("ALIGN", (0,0), (-1,0), "CENTER"),

        ("VALIGN", (0,0), (-1,-1), "TOP"),
        ("GRID", (0,0), (-1,-1), 0.4, colors.grey),

        ("LEFTPADDING", (0,0), (-1,-1), 5),
        ("RIGHTPADDING", (0,0), (-1,-1), 5),
        ("TOPPADDING", (0,0), (-1,-1), 2),
        ("BOTTOMPADDING", (0,0), (-1,-1), 3),
    ]))

    doc.build([tbl])


# -------------------- MAIN --------------------

def main():
    sess = session_with_retries()

    # 1) Aggregate projects across all keywords (dedupe by (id, publicationId))
    dedup: "OrderedDict[Tuple[str,str], Dict]" = OrderedDict()
    title_cache: Dict[str, str] = {}

    for i, kw in enumerate(KEYWORDS, 1):
        try:
            hits = search_projects(sess, kw, CPV_CODES)
        except Exception as e:
            print(f"[warn] search failed for '{kw}': {e}", file=sys.stderr)
            continue

        time.sleep(0.3)  # be gentle to API

        for p in hits:
            pid = p.get("id")
            pub = p.get("publicationId")
            if not pid or not pub:
                continue

            if p.get("title"):
                title_cache[pid] = pick(p.get("title"))  # list-level fallback

            dedup[(pid, pub)] = p  # last write wins (order preserved)

        print(f"[info] {i:>3}/{len(KEYWORDS)} '{kw}': +{len(hits)} (unique so far: {len(dedup)})")

    if not dedup:
        print("[info] No projects found with current keywords/CPVs.")
        return

    # 2) Fetch details for each unique (pid, pub)
    records: List[Dict[str, str]] = []
    for j, ((pid, pub), _) in enumerate(dedup.items(), 1):
        try:
            detail = fetch_detail(sess, pid, pub)
        except Exception as e:
            print(f"[warn] detail failed for {pid}/{pub}: {e}", file=sys.stderr)
            detail = {}

        list_title = title_cache.get(pid, "")
        fields = extract_fields(detail, list_title)
        record = {
            "pid": pid or "",
            "pub": pub or "",
            **fields
        }
        records.append(record)

        if j % 5 == 0:
            time.sleep(0.5)

    # 3) Save CSV + PDF (CSV keeps full text; PDF uses truncated desc)
    csv_path = "tenders_details.csv"
    pdf_path = "tenders_details.pdf"
    write_csv(records, csv_path)
    write_pdf(records, pdf_path)

    print(f"Saved: {csv_path}, {pdf_path}")
    print(f"Total rows: {len(records)}")

if __name__ == "__main__":
    main()
```

```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-

"""
SIMAP tender fetcher (with pagination + past tenders)
- Rolls through pages using 'lastItem'
- Supports date range + extra filters to include past tenders
- Iterates over multilingual keyword list + CPV filters
- Dedupes across keywords and pages
- Exports CSV (full text) + PDF (truncated desc to prevent LayoutError)

Requires:
    pip install requests reportlab
"""

import csv
import re
import html
import time
import sys
from collections import OrderedDict
from typing import Dict, Iterable, List, Tuple, Optional

import requests
from requests.adapters import HTTPAdapter, Retry

from xml.sax.saxutils import escape
from reportlab.platypus import SimpleDocTemplate, Table, TableStyle, Paragraph
from reportlab.lib.pagesizes import A4, landscape
from reportlab.lib.styles import getSampleStyleSheet, ParagraphStyle
from reportlab.lib import colors

# -------------------- CONFIG --------------------

SEARCH_ENDPOINT = "https://prod.simap.ch/api/publications/v2/project/project-search"
DETAIL_ENDPOINT = "https://prod.simap.ch/api/publications/v1/project/{pid}/publication-details/{pub}"

# Keywords to iterate (edit freely)
KEYWORDS: List[str] = [
    "study","studie","étude","etude","studio",
    "analysis","analyse","analisi",
    "economy","economics","wirtschaft","ökonomie","économie","economie","economia",
    "benchmark","benchmarking",
    "wirtschaftsberatung","consulting","conseil","consulenza",
    "it-dienste",
    "beratung, software-entwicklung, internet und hilfestellung",
    "markt- und wirtschaftsforschung","umfragen und statistiken",
    "dienstleistung","service","prestation","prestazione",
    "ausschreibung","appel d'offres","appels d'offres","gara","procurement",
    "offenes verfahren","procédure ouverte","procedure ouverte","procedura aperta",
    "bks","bundesamt für statistik","office fédéral de la statistique",
    "office federal de la statistique","ufficio federale di statistica",
    "federal statistical office","ofs","bfs",
    "seco","staatssekretariat für wirtschaft",
    "secrétariat d'état à l'économie","secretariat d'etat a l'economie",
    "segretariato di stato dell'economia",
    "kanton zürich","kanton zuerich","canton de zurich","cantone zurigo",
    "kanton luzern","canton de lucerne","cantone lucerna",
    "kanton","canton","cantone",
    "region","regionen","région","regione",
    "bundesamt für","bundesamt fuer",
    "bundesamt für bauten und logistik","bundesamt fuer bauten und logistik","bbl",
    "index","verbraucherindex","price index","indice des prix","indice dei prezzi",
    "wirtschaftsforschung","economic research","recherche économique","recherche economique","ricerca economica",
    "tunnel",
]

# Filters (edit as needed)
CPV_CODES = ["72000000", "79300000"]

# Include past tenders using date range (YYYY-MM-DD). Leave None to omit.
NEWEST_PUBLICATION_FROM: Optional[str] = "2019-01-01"   # start date for "past" tenders
NEWEST_PUBLICATION_UNTIL: Optional[str] = None          # e.g. "2025-10-01"

# Optional extra filters (set to [] or None to skip)
LANGS = ["de", "en", "fr", "it"]  # which translated fields to search in
PROCESS_TYPES = []  # ["open","selective","invitation","direct","no_process"]
NEWEST_PUB_TYPES = []  # e.g. ["tender","award_tender","study_contract","award_study_contract"]
PROJECT_SUBTYPES = []  # ["construction","service","supply","project_study",...]
ORDER_ADDRESS_COUNTRY_ONLY_SWITZERLAND: Optional[bool] = None  # True/False/None
ORDER_ADDRESS_CANTONS: List[str] = []  # e.g. ["ZH","BE","VD"]

# Pagination control
PAGE_SLEEP = 0.25  # seconds between page fetches

# PDF safeguards (avoid LayoutError)
MAX_DESC_CHARS = 700       # PDF ONLY (CSV keeps full text)
PDF_FONT_SIZE  = 8
PDF_LEADING    = 10

# PDF page + layout
PAGE_SIZE     = landscape(A4)
MARGINS       = dict(leftMargin=24, rightMargin=24, topMargin=24, bottomMargin=24)
COLUMN_WIDTHS = [120, 120, 80, 220, 300, 140, 100, 180]  # tune if needed


# -------------------- HELPERS --------------------

def pick(d: Dict, langs: Iterable[str] = ("de","en","fr","it")) -> str:
    return next((d.get(k) for k in langs if isinstance(d, dict) and d.get(k)), "") if isinstance(d, dict) else ""

def clean_html(s):
    s = "" if s is None else str(s)
    s = html.unescape(s)
    s = re.sub(r'</p\s*>', '\n', s, flags=re.I)
    s = re.sub(r'<\s*br\s*/?>', '\n', s, flags=re.I)
    s = re.sub(r'<[^>]+>', '', s)
    return s.strip()

def session_with_retries() -> requests.Session:
    s = requests.Session()
    retries = Retry(total=4, backoff_factor=0.4, status_forcelist=(429, 500, 502, 503, 504))
    s.mount("https://", HTTPAdapter(max_retries=retries))
    s.headers.update({"User-Agent": "SIMAPFetcher/1.1 (+https://example.org)"})
    return s

def build_search_params(
    keyword: str,
    cpv_codes: List[str],
    last_item: Optional[str] = None,
) -> Dict:
    params = {
        "search": keyword,
        "cpvCodes": cpv_codes,
    }
    if LANGS:
        params["lang"] = LANGS
    if PROCESS_TYPES:
        params["processTypes"] = PROCESS_TYPES
    if NEWEST_PUB_TYPES:
        params["newestPubTypes"] = NEWEST_PUB_TYPES
    if PROJECT_SUBTYPES:
        params["projectSubTypes"] = PROJECT_SUBTYPES
    if ORDER_ADDRESS_COUNTRY_ONLY_SWITZERLAND is not None:
        params["orderAddressCountryOnlySwitzerland"] = str(ORDER_ADDRESS_COUNTRY_ONLY_SWITZERLAND).lower()
    if ORDER_ADDRESS_CANTONS:
        params["orderAddressCantons"] = ORDER_ADDRESS_CANTONS
    if NEWEST_PUBLICATION_FROM:
        params["newestPublicationFrom"] = NEWEST_PUBLICATION_FROM
    if NEWEST_PUBLICATION_UNTIL:
        params["newestPublicationUntil"] = NEWEST_PUBLICATION_UNTIL
    if last_item:
        params["lastItem"] = last_item
    return params

def search_all_pages(sess: requests.Session, keyword: str, cpv_codes: List[str]) -> List[Dict]:
    """Fetch all pages using rolling pagination with 'lastItem'."""
    all_projects: List[Dict] = []
    last_item: Optional[str] = None
    page = 0
    while True:
        page += 1
        r = sess.get(SEARCH_ENDPOINT, params=build_search_params(keyword, cpv_codes, last_item), timeout=30)
        r.raise_for_status()
        data = r.json() or {}

        projects = data.get("projects", []) or []
        if not projects:
            break

        all_projects.extend(projects)

        # Pagination key is in 'pagination' section, usually 'lastItem' (string like "2025-09-05|12242-06")
        pagination = data.get("pagination") or {}
        next_last_item = pagination.get("lastItem")
        print(f"[info] '{keyword}' page {page}: +{len(projects)} items, next lastItem={next_last_item}")
        if not next_last_item or next_last_item == last_item:
            break
        last_item = next_last_item
        time.sleep(PAGE_SLEEP)

    return all_projects

def fetch_detail(sess: requests.Session, pid: str, pub: str) -> Dict:
    url = DETAIL_ENDPOINT.format(pid=pid, pub=pub)
    r = sess.get(url, timeout=30)
    r.raise_for_status()
    return r.json() or {}

def extract_fields(detail: Dict, list_level_title: str = "") -> Dict[str, str]:
    pr = detail.get("procurement", {}) or {}

    title = pick(pr.get("projectTitle")) or pick(detail.get("title")) or list_level_title or ""
    desc_full = clean_html(pick(pr.get("orderDescription")))
    if len(desc_full) > MAX_DESC_CHARS:
        desc_pdf = desc_full[:MAX_DESC_CHARS].rstrip() + " … [truncated]"
    else:
        desc_pdf = desc_full

    cpv_label = pick((pr.get("cpvCode") or {}).get("label"), ("en","de","fr","it"))
    process   = pr.get("processType", "") or ""
    pub_date  = detail.get("publicationDate") or ""
    office    = pick(pr.get("procOfficeName"), ("de","en","fr","it")) or pick(detail.get("procOfficeName"))

    return dict(
        title=title,
        desc_full=desc_full,
        desc_pdf=desc_pdf,
        cpv_label=cpv_label or "",
        process=process,
        pub_date=pub_date or "",
        office=office or "",
    )

def write_csv(records: List[Dict[str, str]], path: str) -> None:
    header = ["Project ID", "Publication ID", "Process Type", "Title", "Order Description",
              "CPV Label", "Publication Date", "Procurement Office"]
    with open(path, "w", newline="", encoding="utf-8") as f:
        w = csv.writer(f)
        w.writerow(header)
        for r in records:
            w.writerow([
                r["pid"], r["pub"], r["process"], r["title"],
                r["desc_full"], r["cpv_label"], r["pub_date"], r["office"]
            ])

def write_pdf(records: List[Dict[str, str]], path: str) -> None:
    styles = getSampleStyleSheet()
    tiny = ParagraphStyle(
        "Tiny",
        parent=styles["Normal"],
        fontSize=PDF_FONT_SIZE,
        leading=PDF_LEADING,
        wordWrap="CJK",
    )

    def cell(x: str) -> Paragraph:
        return Paragraph(escape(x or ""), tiny)

    header = ["Project ID", "Publication ID", "Process Type", "Title", "Order Description",
              "CPV Label", "Publication Date", "Procurement Office"]
    table_data = [header]
    for r in records:
        table_data.append([
            cell(r["pid"]), cell(r["pub"]), cell(r["process"]), cell(r["title"]),
            cell(r["desc_pdf"]), cell(r["cpv_label"]), cell(r["pub_date"]), cell(r["office"])
        ])

    doc = SimpleDocTemplate(path, pagesize=PAGE_SIZE, **MARGINS, title="SIMAP Tenders")
    tbl = Table(table_data, repeatRows=1, colWidths=COLUMN_WIDTHS)
    tbl.setStyle(TableStyle([
        ("BACKGROUND", (0,0), (-1,0), colors.black),
        ("TEXTCOLOR", (0,0), (-1,0), colors.whitesmoke),
        ("FONTNAME", (0,0), (-1,0), "Helvetica-Bold"),
        ("FONTSIZE", (0,0), (-1,0), 9),
        ("ALIGN", (0,0), (-1,0), "CENTER"),
        ("VALIGN", (0,0), (-1,-1), "TOP"),
        ("GRID", (0,0), (-1,-1), 0.4, colors.grey),
        ("LEFTPADDING", (0,0), (-1,-1), 5),
        ("RIGHTPADDING", (0,0), (-1,-1), 5),
        ("TOPPADDING", (0,0), (-1,-1), 2),
        ("BOTTOMPADDING", (0,0), (-1,-1), 3),
    ]))
    doc.build([tbl])


# -------------------- MAIN --------------------

def main():
    sess = session_with_retries()

    # 1) Aggregate (dedupe by (id, publicationId)) across keywords + pages
    dedup: "OrderedDict[Tuple[str,str], Dict]" = OrderedDict()
    title_cache: Dict[str, str] = {}

    for i, kw in enumerate(KEYWORDS, 1):
        try:
            hits = search_all_pages(sess, kw, CPV_CODES)
        except Exception as e:
            print(f"[warn] search failed for '{kw}': {e}", file=sys.stderr)
            continue

        for p in hits:
            pid = p.get("id")
            pub = p.get("publicationId")
            if not pid or not pub:
                continue
            if p.get("title"):
                title_cache[pid] = pick(p.get("title"))
            dedup[(pid, pub)] = p

        print(f"[info] {i:>3}/{len(KEYWORDS)} '{kw}': total+{len(hits)} (unique so far: {len(dedup)})")

    if not dedup:
        print("[info] No projects found with current filters.")
        return

    # 2) Fetch details
    records: List[Dict[str, str]] = []
    for j, ((pid, pub), _) in enumerate(dedup.items(), 1):
        try:
            detail = fetch_detail(sess, pid, pub)
        except Exception as e:
            print(f"[warn] detail failed for {pid}/{pub}: {e}", file=sys.stderr)
            detail = {}
        fields = extract_fields(detail, title_cache.get(pid, ""))
        record = {"pid": pid or "", "pub": pub or "", **fields}
        records.append(record)
        if j % 5 == 0:
            time.sleep(0.4)

    # 3) Save
    csv_path = "tenders_details.csv"
    pdf_path = "tenders_details.pdf"
    write_csv(records, csv_path)
    write_pdf(records, pdf_path)

    print(f"Saved: {csv_path}, {pdf_path}")
    print(f"Total rows: {len(records)}")

if __name__ == "__main__":
    main()
```

```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-

"""
SIMAP tender fetcher (single file)
- Iterates over a multilingual keyword list + CPV filters
- Filters by publication date: 2018-01-01 .. 2024-12-12
- Dedupes results across keywords
- Fetches detail JSON per project/publication
- Exports:
    * CSV  -> keeps FULL descriptions
    * PDF  -> truncates long descriptions to avoid ReportLab LayoutError
- Publication Date column now falls back to the list-level date when detail lacks it.
- PDF is A4 landscape with smaller font and comfy margins

Requires:
    pip install requests reportlab
"""

import csv
import re
import html
import time
import sys
from collections import OrderedDict
from typing import Dict, Iterable, List, Tuple

import requests
from requests.adapters import HTTPAdapter, Retry

from xml.sax.saxutils import escape
from reportlab.platypus import SimpleDocTemplate, Table, TableStyle, Paragraph
from reportlab.lib.pagesizes import A4, landscape
from reportlab.lib.styles import getSampleStyleSheet, ParagraphStyle
from reportlab.lib import colors

# -------------------- CONFIG --------------------

SEARCH_ENDPOINT = "https://prod.simap.ch/api/publications/v2/project/project-search"
DETAIL_ENDPOINT = "https://prod.simap.ch/api/publications/v1/project/{pid}/publication-details/{pub}"

# Multilingual keywords (edit freely)
KEYWORDS: List[str] = [
    "study","studie","étude","etude","studio",
    "analysis","analyse","analisi",
    "economy","economics","wirtschaft","ökonomie","économie","economie","economia",
    "benchmark","benchmarking",
    "wirtschaftsberatung","consulting","conseil","consulenza",
    "72000000","79300000",
    "it-dienste",
    "beratung, software-entwicklung, internet und hilfestellung",
    "markt- und wirtschaftsforschung","umfragen und statistiken",
    "dienstleistung","service","prestation","prestazione",
    "ausschreibung","appel d'offres","appels d'offres","gara","procurement",
    "offenes verfahren","procédure ouverte","procedure ouverte","procedura aperta",
    "bks","bundesamt für statistik","office fédéral de la statistique",
    "office federal de la statistique","ufficio federale di statistica",
    "federal statistical office","ofs","bfs",
    "seco","staatssekretariat für wirtschaft",
    "secrétariat d'état à l'économie","secretariat d'etat a l'economie",
    "segretariato di stato dell'economia",
    "kanton zürich","kanton zuerich","canton de zurich","cantone zurigo",
    "kanton luzern","canton de lucerne","cantone lucerna",
    "kanton","canton","cantone",
    "region","regionen","région","regione",
    "bundesamt für","bundesamt fuer",
    "bundesamt für bauten und logistik","bundesamt fuer bauten und logistik","bbl",
    "index","verbraucherindex","price index","indice des prix","indice dei prezzi",
    "wirtschaftsforschung","economic research","recherche économique","recherche economique","ricerca economica",
    "tunnel",
]

# CPV filter (edit if needed)
CPV_CODES = ["72000000", "79300000"]

# Date filter (as requested)
NEWEST_PUBLICATION_FROM = "2018-01-01"
NEWEST_PUBLICATION_UNTIL = "2024-12-12"

# PDF safeguards (avoid LayoutError)
MAX_DESC_CHARS = 700       # PDF ONLY (CSV keeps full text)
PDF_FONT_SIZE  = 8
PDF_LEADING    = 10

# PDF page + layout
PAGE_SIZE     = landscape(A4)
MARGINS       = dict(leftMargin=24, rightMargin=24, topMargin=24, bottomMargin=24)
COLUMN_WIDTHS = [120, 120, 80, 220, 300, 140, 100, 180]  # tune if needed


# -------------------- HELPERS --------------------

def pick(d: Dict, langs: Iterable[str] = ("de","en","fr","it")) -> str:
    """Pick the first non-empty localized value from a dict like {'de': '...', 'en': None, ...}."""
    return next((d.get(k) for k in langs if isinstance(d, dict) and d.get(k)), "") if isinstance(d, dict) else ""

def clean_html(s):
    s = "" if s is None else str(s)
    s = html.unescape(s)
    s = re.sub(r'</p\s*>', '\n', s, flags=re.I)
    s = re.sub(r'<\s*br\s*/?>', '\n', s, flags=re.I)
    s = re.sub(r'<[^>]+>', '', s)  # strip remaining tags
    return s.strip()

def session_with_retries() -> requests.Session:
    s = requests.Session()
    retries = Retry(total=4, backoff_factor=0.4, status_forcelist=(429, 500, 502, 503, 504))
    s.mount("https://", HTTPAdapter(max_retries=retries))
    s.headers.update({"User-Agent": "SIMAPFetcher/1.0 (+https://example.org)"})
    return s

def search_projects(sess: requests.Session, keyword: str, cpv_codes: List[str]) -> List[Dict]:
    # Added date filters (2018-01-01 .. 2024-12-12)
    params = {
        "search": keyword,
        "cpvCodes": cpv_codes,
        "newestPublicationFrom": NEWEST_PUBLICATION_FROM,
        "newestPublicationUntil": NEWEST_PUBLICATION_UNTIL,
    }
    r = sess.get(SEARCH_ENDPOINT, params=params, timeout=20)
    r.raise_for_status()
    data = r.json() or {}
    return data.get("projects", []) or []

def fetch_detail(sess: requests.Session, pid: str, pub: str) -> Dict:
    url = DETAIL_ENDPOINT.format(pid=pid, pub=pub)
    r = sess.get(url, timeout=30)
    r.raise_for_status()
    return r.json() or {}

def extract_fields(detail: Dict, list_level_title: str = "", fallback_pub_date: str = "") -> Dict[str, str]:
    """Return all fields + both full and truncated descriptions (for CSV/PDF).
       Uses list-level publication date as fallback when detail lacks it."""
    pr = detail.get("procurement", {}) or {}

    title = pick(pr.get("projectTitle")) or pick(detail.get("title")) or list_level_title or ""
    desc_full = clean_html(pick(pr.get("orderDescription")))  # CSV uses full
    # PDF truncated version
    desc_pdf = (desc_full[:MAX_DESC_CHARS].rstrip() + " … [truncated]") if len(desc_full) > MAX_DESC_CHARS else desc_full

    cpv_label = pick((pr.get("cpvCode") or {}).get("label"), ("en","de","fr","it"))
    process   = pr.get("processType", "") or ""
    pub_date  = detail.get("publicationDate") or fallback_pub_date or ""  # <-- fallback here
    office    = pick(pr.get("procOfficeName"), ("de","en","fr","it")) or pick(detail.get("procOfficeName"))

    return dict(
        title=title,
        desc_full=desc_full,
        desc_pdf=desc_pdf,
        cpv_label=cpv_label or "",
        process=process,
        pub_date=pub_date,
        office=office or "",
    )

def write_csv(records: List[Dict[str, str]], path: str) -> None:
    header = ["Project ID", "Publication ID", "Process Type", "Title", "Order Description",
              "CPV Label", "Publication Date", "Procurement Office"]
    with open(path, "w", newline="", encoding="utf-8") as f:
        w = csv.writer(f)
        w.writerow(header)
        for r in records:
            w.writerow([
                r["pid"], r["pub"], r["process"], r["title"],
                r["desc_full"], r["cpv_label"], r["pub_date"], r["office"]
            ])

def write_pdf(records: List[Dict[str, str]], path: str) -> None:
    styles = getSampleStyleSheet()
    tiny = ParagraphStyle(
        "Tiny",
        parent=styles["Normal"],
        fontSize=PDF_FONT_SIZE,
        leading=PDF_LEADING,
        wordWrap="CJK",  # better wrapping for long tokens
    )

    def cell(x: str) -> Paragraph:
        return Paragraph(escape(x or ""), tiny)

    header = ["Project ID", "Publication ID", "Process Type", "Title", "Order Description",
              "CPV Label", "Publication Date", "Procurement Office"]
    table_data = [header]
    for r in records:
        table_data.append([
            cell(r["pid"]), cell(r["pub"]), cell(r["process"]), cell(r["title"]),
            cell(r["desc_pdf"]), cell(r["cpv_label"]), cell(r["pub_date"]), cell(r["office"])
        ])

    doc = SimpleDocTemplate(
        path, pagesize=PAGE_SIZE,
        **MARGINS, title="SIMAP Tenders"
    )

    tbl = Table(table_data, repeatRows=1, colWidths=COLUMN_WIDTHS)
    tbl.setStyle(TableStyle([
        ("BACKGROUND", (0,0), (-1,0), colors.black),
        ("TEXTCOLOR", (0,0), (-1,0), colors.whitesmoke),
        ("FONTNAME", (0,0), (-1,0), "Helvetica-Bold"),
        ("FONTSIZE", (0,0), (-1,0), 9),
        ("ALIGN", (0,0), (-1,0), "CENTER"),

        ("VALIGN", (0,0), (-1,-1), "TOP"),
        ("GRID", (0,0), (-1,-1), 0.4, colors.grey),

        ("LEFTPADDING", (0,0), (-1,-1), 5),
        ("RIGHTPADDING", (0,0), (-1,-1), 5),
        ("TOPPADDING", (0,0), (-1,-1), 2),
        ("BOTTOMPADDING", (0,0), (-1,-1), 3),
    ]))

    doc.build([tbl])


# -------------------- MAIN --------------------

def main():
    sess = session_with_retries()

    # 1) Aggregate projects across all keywords (dedupe by (id, publicationId))
    dedup: "OrderedDict[Tuple[str,str], Dict]" = OrderedDict()
    title_cache: Dict[str, str] = {}

    for i, kw in enumerate(KEYWORDS, 1):
        try:
            hits = search_projects(sess, kw, CPV_CODES)
        except Exception as e:
            print(f"[warn] search failed for '{kw}': {e}", file=sys.stderr)
            continue

        time.sleep(0.3)  # be gentle to API

        for p in hits:
            pid = p.get("id")
            pub = p.get("publicationId")
            if not pid or not pub:
                continue

            if p.get("title"):
                title_cache[pid] = pick(p.get("title"))  # list-level fallback

            dedup[(pid, pub)] = p  # last write wins (order preserved)

        print(f"[info] {i:>3}/{len(KEYWORDS)} '{kw}': +{len(hits)} (unique so far: {len(dedup)})")

    if not dedup:
        print("[info] No projects found with current keywords/CPVs.")
        return

    # 2) Fetch details for each unique (pid, pub)
    records: List[Dict[str, str]] = []
    for j, ((pid, pub), p) in enumerate(dedup.items(), 1):  # keep list item `p` to get fallback date
        try:
            detail = fetch_detail(sess, pid, pub)
        except Exception as e:
            print(f"[warn] detail failed for {pid}/{pub}: {e}", file=sys.stderr)
            detail = {}

        list_title     = title_cache.get(pid, "")
        list_pub_date  = p.get("publicationDate") or ""  # fallback date from list item
        fields = extract_fields(detail, list_title, fallback_pub_date=list_pub_date)
        record = {
            "pid": pid or "",
            "pub": pub or "",
            **fields
        }
        records.append(record)

        if j % 5 == 0:
            time.sleep(0.5)

    # 3) Save CSV + PDF (CSV keeps full text; PDF uses truncated desc)
    csv_path = "tenders_details.csv"
    pdf_path = "tenders_details.pdf"
    write_csv(records, csv_path)
    write_pdf(records, pdf_path)

    print(f"Saved: {csv_path}, {pdf_path}")
    print(f"Total rows: {len(records)}")

if __name__ == "__main__":
    main()
```

```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-

# pip install requests reportlab

import csv, re, html, time, sys
from collections import OrderedDict
from typing import Dict, Iterable, List, Tuple, Optional

import requests
from requests.adapters import HTTPAdapter, Retry

from xml.sax.saxutils import escape
from reportlab.platypus import SimpleDocTemplate, Table, TableStyle, Paragraph
from reportlab.lib.pagesizes import A4, landscape
from reportlab.lib.styles import getSampleStyleSheet, ParagraphStyle
from reportlab.lib import colors

SEARCH_ENDPOINT = "https://prod.simap.ch/api/publications/v2/project/project-search"
DETAIL_ENDPOINT = "https://prod.simap.ch/api/publications/v1/project/{pid}/publication-details/{pub}"

# ---- Zorunlu tarih aralığı (DAHİL): 2018-01-01 .. 2024-12-12 ----
NEWEST_PUBLICATION_FROM  = "2018-01-01"
NEWEST_PUBLICATION_UNTIL = "2024-12-12"

# ---- Opsiyonel anahtar kelimeler (boş da bırakabilirsiniz) ----
KEYWORDS: List[str] = [
    "study","studie","étude","analysis","economy","benchmark",
    "wirtschaftsberatung","consulting","it-dienste","dienstleistung",
    "ausschreibung","procurement","offenes verfahren","bks","seco",
    "kanton","region","bundesamt","index","wirtschaftsforschung","tunnel"
]

# Ek filtre isterseniz (tercihe bağlı)
CPV_CODES = ["72000000","79300000"]  # boş liste yapabilirsiniz: []

# Performans
PAGE_SLEEP = 0.20          # sayfalar arası bekleme
DETAIL_SLEEP = 0.40        # her 5 detayda bir ufak bekleme
MAX_PAGES_PER_QUERY = 0    # 0 = sınırsız; >0 = her sorgu için sayfa limiti

# PDF güvenli kısaltma
MAX_DESC_CHARS = 700
PDF_FONT_SIZE  = 8
PDF_LEADING    = 10

PAGE_SIZE     = landscape(A4)
MARGINS       = dict(leftMargin=24, rightMargin=24, topMargin=24, bottomMargin=24)
COLUMN_WIDTHS = [120, 120, 80, 220, 300, 140, 100, 180]

# ---------- yardımcılar ----------
def pick(d: Dict, langs: Iterable[str] = ("de","en","fr","it")) -> str:
    return next((d.get(k) for k in langs if isinstance(d, dict) and d.get(k)), "") if isinstance(d, dict) else ""

def clean_html(s):
    s = "" if s is None else str(s)
    s = html.unescape(s)
    s = re.sub(r'</p\s*>', '\n', s, flags=re.I)
    s = re.sub(r'<\s*br\s*/?>', '\n', s, flags=re.I)
    s = re.sub(r'<[^>]+>', '', s)
    return s.strip()

def session_with_retries() -> requests.Session:
    s = requests.Session()
    retries = Retry(total=4, backoff_factor=0.4, status_forcelist=(429,500,502,503,504))
    s.mount("https://", HTTPAdapter(max_retries=retries))
    s.headers.update({"User-Agent": "SIMAPFetcher/1.2 (+https://example.org)"})
    return s

def build_params(search_value: Optional[str], last_item: Optional[str]=None) -> Dict:
    params = {
        "newestPublicationFrom":  NEWEST_PUBLICATION_FROM,
        "newestPublicationUntil": NEWEST_PUBLICATION_UNTIL,
        "lang": ["de","en","fr","it"],
    }
    if CPV_CODES:
        params["cpvCodes"] = CPV_CODES
    if search_value is not None:
        params["search"] = search_value
    if last_item:
        params["lastItem"] = last_item
    return params

def fetch_page(sess: requests.Session, params: Dict) -> Dict:
    r = sess.get(SEARCH_ENDPOINT, params=params, timeout=30)
    r.raise_for_status()
    return r.json() or {}

def search_all_pages(sess: requests.Session, search_value: Optional[str]) -> List[Dict]:
    """Rolling pagination. search_value None ise önce aramasız dener, 400 gelirse 'und' ile dener."""
    all_items: List[Dict] = []
    last_item, page = None, 0
    while True:
        if MAX_PAGES_PER_QUERY and page >= MAX_PAGES_PER_QUERY:
            break
        page += 1
        params = build_params(search_value, last_item)
        try:
            data = fetch_page(sess, params)
        except requests.HTTPError as e:
            # Aramasız istek (search=None) 400 verirse, 'und' ile tekrar dene
            if search_value is None:
                search_value = "und"  # nötr ve yaygın Almanca bağlaç
                data = fetch_page(sess, build_params(search_value, last_item))
            else:
                raise e
        batch = data.get("projects", []) or []
        if not batch:
            break
        all_items.extend(batch)
        pag = data.get("pagination") or {}
        next_last_item = pag.get("lastItem")
        if not next_last_item or next_last_item == last_item:
            break
        last_item = next_last_item
        time.sleep(PAGE_SLEEP)
    return all_items

def fetch_detail(sess: requests.Session, pid: str, pub: str) -> Dict:
    url = DETAIL_ENDPOINT.format(pid=pid, pub=pub)
    r = sess.get(url, timeout=30)
    r.raise_for_status()
    return r.json() or {}

def extract_fields(detail: Dict, list_level_title: str = "", fallback_pub_date: str = "") -> Dict[str,str]:
    pr = detail.get("procurement", {}) or {}
    title = pick(pr.get("projectTitle")) or pick(detail.get("title")) or list_level_title or ""
    desc_full = clean_html(pick(pr.get("orderDescription")))
    desc_pdf  = (desc_full[:MAX_DESC_CHARS].rstrip() + " … [truncated]") if len(desc_full) > MAX_DESC_CHARS else desc_full
    cpv_label = pick((pr.get("cpvCode") or {}).get("label"), ("en","de","fr","it"))
    process   = pr.get("processType", "") or ""
    pub_date  = detail.get("publicationDate") or fallback_pub_date or ""
    office    = pick(pr.get("procOfficeName"), ("de","en","fr","it")) or pick(detail.get("procOfficeName"))
    return dict(
        title=title, desc_full=desc_full, desc_pdf=desc_pdf,
        cpv_label=cpv_label or "", process=process, pub_date=pub_date, office=office or ""
    )

def write_csv(records: List[Dict[str,str]], path: str) -> None:
    header = ["Project ID","Publication ID","Process Type","Title","Order Description",
              "CPV Label","Publication Date","Procurement Office"]
    with open(path, "w", newline="", encoding="utf-8") as f:
        w = csv.writer(f); w.writerow(header)
        for r in records:
            w.writerow([r["pid"], r["pub"], r["process"], r["title"],
                        r["desc_full"], r["cpv_label"], r["pub_date"], r["office"]])

def write_pdf(records: List[Dict[str,str]], path: str) -> None:
    styles = getSampleStyleSheet()
    tiny = ParagraphStyle("Tiny", parent=styles["Normal"],
                          fontSize=PDF_FONT_SIZE, leading=PDF_LEADING, wordWrap="CJK")
    def cell(x:str)->Paragraph: return Paragraph(escape(x or ""), tiny)
    header = ["Project ID","Publication ID","Process Type","Title","Order Description",
              "CPV Label","Publication Date","Procurement Office"]
    table_data = [header] + [[cell(r[k]) for k in
                     ("pid","pub","process","title","desc_pdf","cpv_label","pub_date","office")]
                    for r in records]
    doc = SimpleDocTemplate(path, pagesize=PAGE_SIZE, title="SIMAP Tenders", **MARGINS)
    tbl = Table(table_data, repeatRows=1, colWidths=COLUMN_WIDTHS)
    tbl.setStyle(TableStyle([
        ("BACKGROUND",(0,0),(-1,0),colors.black),
        ("TEXTCOLOR",(0,0),(-1,0),colors.whitesmoke),
        ("FONTNAME",(0,0),(-1,0),"Helvetica-Bold"),
        ("FONTSIZE",(0,0),(-1,0),9),
        ("ALIGN",(0,0),(-1,0),"CENTER"),
        ("VALIGN",(0,0),(-1,-1),"TOP"),
        ("GRID",(0,0),(-1,-1),0.4,colors.grey),
        ("LEFTPADDING",(0,0),(-1,-1),5),
        ("RIGHTPADDING",(0,0),(-1,-1),5),
        ("TOPPADDING",(0,0),(-1,-1),2),
        ("BOTTOMPADDING",(0,0),(-1,-1),3),
    ]))
    doc.build([tbl])

# ---------- ana akış ----------
def main():
    sess = session_with_retries()

    # 1) Anahtar kelimelerle tara
    dedup: "OrderedDict[Tuple[str,str], Dict]" = OrderedDict()
    title_cache: Dict[str,str] = {}
    print(f"[info] Tarih filtresi: {NEWEST_PUBLICATION_FROM} .. {NEWEST_PUBLICATION_UNTIL}")

    for i, kw in enumerate(KEYWORDS, 1):
        kw = kw.strip()
        if not kw or len(kw) < 3: continue
        hits = search_all_pages(sess, kw)
        for p in hits:
            pid, pub = p.get("id"), p.get("publicationId")
            if not pid or not pub: continue
            if p.get("title"): title_cache[pid] = pick(p.get("title"))
            dedup[(pid, pub)] = p
        print(f"[info] {i:>3}/{len(KEYWORDS)} '{kw}': +{len(hits)} (toplam uniq: {len(dedup)})")

    # 2) Yedek tarama (keyword olmadan ya da 'und' ile) — 2018 kayıtlarını kaçırmamak için
    backup_hits = search_all_pages(sess, search_value=None)
    for p in backup_hits:
        pid, pub = p.get("id"), p.get("publicationId")
        if not pid or not pub: continue
        if p.get("title"): title_cache[pid] = pick(p.get("title"))
        dedup[(pid, pub)] = p
    print(f"[info] backup tarama: +{len(backup_hits)} (toplam uniq: {len(dedup)})")

    if not dedup:
        print("[info] Kayıt bulunamadı. Filtreleri/kelimeleri gevşetmeyi deneyin.")
        return

    # 3) Detayları çek ve kayıtları hazırla
    records: List[Dict[str,str]] = []
    for j, ((pid, pub), p) in enumerate(dedup.items(), 1):
        try:
            detail = fetch_detail(sess, pid, pub)
        except Exception as e:
            print(f"[warn] detail {pid}/{pub}: {e}", file=sys.stderr)
            detail = {}
        list_title    = title_cache.get(pid, "")
        list_pub_date = p.get("publicationDate") or ""
        fields = extract_fields(detail, list_title, fallback_pub_date=list_pub_date)
        record = {"pid": pid or "", "pub": pub or "", **fields}
        records.append(record)
        if j % 5 == 0:
            time.sleep(DETAIL_SLEEP)

    # 4) Dışa aktar
    write_csv(records, "tenders_2018_2024.csv")
    write_pdf(records, "tenders_2018_2024.pdf")
    print(f"[done] CSV+PDF hazır: {len(records)} satır")

if __name__ == "__main__":
    main()
```

```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-

"""
SIMAP tender fetcher (pagination'lı, tek dosya)
- Zorunlu tarih filtresi: 2018-01-01 .. 2024-12-12 (dahil)
- Rolling pagination: 'lastItem' ile tüm sayfaları gezer
- (İsteğe bağlı) çok dilli anahtar kelimeler + CPV filtreleri
- Dedup: (project id, publication id) bazında tekilleştirir
- Detay endpoint'ten açıklamayı çeker; CSV tam metin, PDF kısaltılmış
- PDF: A4 landscape, küçük font; uzun hücrelerde LayoutError önlenir

Gereken paketler:
    pip install requests reportlab
"""

import csv, re, html, time, sys
from collections import OrderedDict
from typing import Dict, Iterable, List, Tuple, Optional

import requests
from requests.adapters import HTTPAdapter, Retry

from xml.sax.saxutils import escape
from reportlab.platypus import SimpleDocTemplate, Table, TableStyle, Paragraph
from reportlab.lib.pagesizes import A4, landscape
from reportlab.lib.styles import getSampleStyleSheet, ParagraphStyle
from reportlab.lib import colors

# -------------------- SABİTLER / AYARLAR --------------------

SEARCH_ENDPOINT = "https://prod.simap.ch/api/publications/v2/project/project-search"
DETAIL_ENDPOINT = "https://prod.simap.ch/api/publications/v1/project/{pid}/publication-details/{pub}"

# Zorunlu tarih aralığı (DAHİL)
NEWEST_PUBLICATION_FROM  = "2018-01-01"
NEWEST_PUBLICATION_UNTIL = "2019-01-01"

# (Opsiyonel) anahtar kelimeler — boş bırakabilirsiniz
KEYWORDS: List[str] = [
    "study","studie","étude","analysis","economy","benchmark",
    "wirtschaftsberatung","consulting","it-dienste","dienstleistung",
    "ausschreibung","procurement","offenes verfahren","bks","seco",
    "kanton","region","bundesamt","index","wirtschaftsforschung","tunnel"
]

# (Opsiyonel) CPV filtreleri — genişletmek için [] yapabilirsinizCPV_CODES: List[str] = ["72000000","79300000"]

# Dil taraması (translated alanlarda arama kapsamı)
LANGS = ["de","en","fr","it"]

# Performans / naziklik
PAGE_SLEEP   = 0.20   # sayfa arası bekleme
DETAIL_SLEEP = 0.40   # her 5 detaydan sonra bekleme
MAX_PAGES_PER_QUERY = 0  # 0 = sınırsız; >0 = sayfa limiti

# PDF güvenlik/kısaltma
MAX_DESC_CHARS = 700
PDF_FONT_SIZE  = 8
PDF_LEADING    = 10
PAGE_SIZE      = landscape(A4)
MARGINS        = dict(leftMargin=24, rightMargin=24, topMargin=24, bottomMargin=24)
COLUMN_WIDTHS  = [120, 120, 80, 220, 300, 140, 100, 180]

# -------------------- YARDIMCI FONKSİYONLAR --------------------

def pick(d: Dict, langs: Iterable[str] = ("de","en","fr","it")) -> str:
    """{'de': '...', 'en': None, ...} tipindeki sözlükten ilk dolu değeri seçer."""
    return next((d.get(k) for k in langs if isinstance(d, dict) and d.get(k)), "") if isinstance(d, dict) else ""

def clean_html(s: Optional[str]) -> str:
    s = "" if s is None else str(s)
    s = html.unescape(s)
    s = re.sub(r'</p\s*>', '\n', s, flags=re.I)
    s = re.sub(r'<\s*br\s*/?>', '\n', s, flags=re.I)
    s = re.sub(r'<[^>]+>', '', s)
    return s.strip()

def session_with_retries() -> requests.Session:
    s = requests.Session()
    retries = Retry(total=4, backoff_factor=0.4, status_forcelist=(429,500,502,503,504))
    s.mount("https://", HTTPAdapter(max_retries=retries))
    s.headers.update({"User-Agent": "SIMAPFetcher/1.3 (+https://example.org)"})
    return s

def build_search_params(search_value: Optional[str], last_item: Optional[str]=None) -> Dict:
    """
    SIMAP gereği: 'search' BOŞ OLAMAZ; ya search >=3 karakter
    ya da bir quick-filter verilmelidir. Biz güvenli olmak için:
    - eğer keyword verilmişse onu koyuyoruz,
    - yedek taramada nötr 'und' kullanıyoruz (çok yaygın Almanca bağlaç).
    """
    params: Dict = {
        "newestPublicationFrom":  NEWEST_PUBLICATION_FROM,
        "newestPublicationUntil": NEWEST_PUBLICATION_UNTIL,
        "lang": LANGS,
    }
    if CPV_CODES:
        params["cpvCodes"] = CPV_CODES
    if last_item:
        params["lastItem"] = last_item

    if search_value is None:
        params["search"] = "und"   # boş veremeyiz; nötr bir arama değeri
    else:
        sv = search_value.strip()
        params["search"] = sv if len(sv) >= 3 else "und"
    return params

def search_all_pages(sess: requests.Session, search_value: Optional[str]) -> List[Dict]:
    """Rolling pagination: pagination.lastItem üzerinden tüm sayfaları dolaşır."""
    all_items: List[Dict] = []
    last_item: Optional[str] = None
    page = 0
    while True:
        if MAX_PAGES_PER_QUERY and page >= MAX_PAGES_PER_QUERY:
            break
        page += 1

        r = sess.get(SEARCH_ENDPOINT, params=build_search_params(search_value, last_item), timeout=30)
        r.raise_for_status()
        data = r.json() or {}

        batch = data.get("projects", []) or []
        if not batch:
            break

        all_items.extend(batch)

        pag = data.get("pagination") or {}
        next_last_item = pag.get("lastItem")
        print(f"[info] '{search_value or 'ALL'}' page {page}: +{len(batch)} items, next={next_last_item}")
        if not next_last_item or next_last_item == last_item:
            break
        last_item = next_last_item
        time.sleep(PAGE_SLEEP)

    return all_items

def fetch_detail(sess: requests.Session, pid: str, pub: str) -> Dict:
    url = DETAIL_ENDPOINT.format(pid=pid, pub=pub)
    r = sess.get(url, timeout=30)
    r.raise_for_status()
    return r.json() or {}

def extract_fields(detail: Dict, list_level_title: str = "", fallback_pub_date: str = "") -> Dict[str, str]:
    """CSV için tam açıklama, PDF için kısaltılmış açıklama üret; tarih için fallback kullan."""
    pr = detail.get("procurement", {}) or {}
    title = pick(pr.get("projectTitle")) or pick(detail.get("title")) or list_level_title or ""
    desc_full = clean_html(pick(pr.get("orderDescription")))
    desc_pdf  = (desc_full[:MAX_DESC_CHARS].rstrip() + " … [truncated]") if len(desc_full) > MAX_DESC_CHARS else desc_full
    cpv_label = pick((pr.get("cpvCode") or {}).get("label"), ("en","de","fr","it"))
    process   = pr.get("processType", "") or ""
    pub_date  = detail.get("publicationDate") or fallback_pub_date or ""
    office    = pick(pr.get("procOfficeName"), ("de","en","fr","it")) or pick(detail.get("procOfficeName"))
    return dict(
        title=title, desc_full=desc_full, desc_pdf=desc_pdf,
        cpv_label=cpv_label or "", process=process, pub_date=pub_date, office=office or ""
    )

def write_csv(records: List[Dict[str, str]], path: str) -> None:
    header = ["Project ID","Publication ID","Process Type","Title","Order Description",
              "CPV Label","Publication Date","Procurement Office"]
    with open(path, "w", newline="", encoding="utf-8") as f:
        w = csv.writer(f)
        w.writerow(header)
        for r in records:
            w.writerow([r["pid"], r["pub"], r["process"], r["title"],
                        r["desc_full"], r["cpv_label"], r["pub_date"], r["office"]])

def write_pdf(records: List[Dict[str, str]], path: str) -> None:
    styles = getSampleStyleSheet()
    tiny = ParagraphStyle("Tiny", parent=styles["Normal"],
                          fontSize=PDF_FONT_SIZE, leading=PDF_LEADING, wordWrap="CJK")
    def cell(x: str) -> Paragraph:
        return Paragraph(escape(x or ""), tiny)

    header = ["Project ID","Publication ID","Process Type","Title","Order Description",
              "CPV Label","Publication Date","Procurement Office"]
    table_data = [header] + [[cell(r[k]) for k in
                     ("pid","pub","process","title","desc_pdf","cpv_label","pub_date","office")]
                    for r in records]

    doc = SimpleDocTemplate(path, pagesize=PAGE_SIZE, title="SIMAP Tenders", **MARGINS)
    tbl = Table(table_data, repeatRows=1, colWidths=COLUMN_WIDTHS)
    tbl.setStyle(TableStyle([
        ("BACKGROUND",(0,0),(-1,0),colors.black),
        ("TEXTCOLOR",(0,0),(-1,0),colors.whitesmoke),
        ("FONTNAME",(0,0),(-1,0),"Helvetica-Bold"),
        ("FONTSIZE",(0,0),(-1,0),9),
        ("ALIGN",(0,0),(-1,0),"CENTER"),
        ("VALIGN",(0,0),(-1,-1),"TOP"),
        ("GRID",(0,0),(-1,-1),0.4,colors.grey),
        ("LEFTPADDING",(0,0),(-1,-1),5),
        ("RIGHTPADDING",(0,0),(-1,-1),5),
        ("TOPPADDING",(0,0),(-1,-1),2),
        ("BOTTOMPADDING",(0,0),(-1,-1),3),
    ]))
    doc.build([tbl])

# -------------------- ANA AKIŞ --------------------

def main():
    sess = session_with_retries()
    print(f"[info] Tarih filtresi: {NEWEST_PUBLICATION_FROM} .. {NEWEST_PUBLICATION_UNTIL}")

    # 1) Keyword bazlı tarama (pagination'lı)
    dedup: "OrderedDict[Tuple[str,str], Dict]" = OrderedDict()
    title_cache: Dict[str, str] = {}

    if KEYWORDS:
        for i, kw in enumerate(KEYWORDS, 1):
            hits = search_all_pages(sess, kw)
            for p in hits:
                pid, pub = p.get("id"), p.get("publicationId")
                if not pid or not pub: continue
                if p.get("title"): title_cache[pid] = pick(p.get("title"))
                dedup[(pid, pub)] = p
            print(f"[info] {i:>3}/{len(KEYWORDS)} '{kw}': +{len(hits)} (uniq: {len(dedup)})")

    # 2) Yedek tarama (çok nötr bir aramayla tüm aralığı süpür)
    backup_hits = search_all_pages(sess, search_value=None)  # None -> 'und' ile arar
    for p in backup_hits:
        pid, pub = p.get("id"), p.get("publicationId")
        if not pid or not pub: continue
        if p.get("title"): title_cache[pid] = pick(p.get("title"))
        dedup[(pid, pub)] = p
    print(f"[info] backup: +{len(backup_hits)} (toplam uniq: {len(dedup)})")

    if not dedup:
        print("[info] Kayıt bulunamadı. Filtreleri gevşetin (CPV, keyword vs.).")
        return

    # 3) Detayları çek ve kayıtları hazırla
    records: List[Dict[str, str]] = []
    for j, ((pid, pub), p) in enumerate(dedup.items(), 1):
        try:
            detail = fetch_detail(sess, pid, pub)
        except Exception as e:
            print(f"[warn] detail {pid}/{pub}: {e}", file=sys.stderr)
            detail = {}

        list_title    = title_cache.get(pid, "")
        list_pub_date = p.get("publicationDate") or ""
        fields = extract_fields(detail, list_title, fallback_pub_date=list_pub_date)
        records.append({"pid": pid or "", "pub": pub or "", **fields})

        if j % 5 == 0:
            time.sleep(DETAIL_SLEEP)

    # 4) Dışa aktar
    write_csv(records, "tenders_2018_2024.csv")
    write_pdf(records, "tenders_2018_2024.pdf")
    print(f"[done] CSV + PDF hazır: {len(records)} satır")

if __name__ == "__main__":
    main()
```

```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-

# pip install requests reportlab

import csv, re, html, time, sys
from collections import OrderedDict
from typing import Dict, Iterable, List, Tuple, Optional

import requests
from requests.adapters import HTTPAdapter, Retry

from xml.sax.saxutils import escape
from reportlab.platypus import SimpleDocTemplate, Table, TableStyle, Paragraph
from reportlab.lib.pagesizes import A4, landscape
from reportlab.lib.styles import getSampleStyleSheet, ParagraphStyle
from reportlab.lib import colors

SEARCH_ENDPOINT = "https://prod.simap.ch/api/publications/v2/project/project-search"
DETAIL_ENDPOINT = "https://prod.simap.ch/api/publications/v1/project/{pid}/publication-details/{pub}"

# --- Zorunlu tarih aralığı (2018 dahil) ---
NEWEST_PUBLICATION_FROM  = "2018-01-01"
NEWEST_PUBLICATION_UNTIL = "2018-12-31"



# (İsteğe bağlı) anahtar kelimeler – ister kullan, ister kullanma
KEYWORDS: List[str] = [
    "study","studie","étude","analysis","economy","benchmark",
    "wirtschaftsberatung","consulting","it-dienste","dienstleistung",
    "ausschreibung","procurement","offenes verfahren","bks","seco",
    "kanton","region","bundesamt","index","wirtschaftsforschung","tunnel"
]

# CPV filtresi: GENİŞ SONUÇ İÇİN ŞİMDİLİK KAPALI ([])
# İstersen tekrar aç: ["72000000","79300000"]
CPV_CODES: List[str] = []

# Hangi alt türleri dahil edelim? (quick-filter; search yerine bunu veriyoruz)
PROJECT_SUBTYPES: List[str] = [
    "construction","service","supply",
    "project_competition","idea_competition","overall_performance_competition",
    "project_study","idea_study","overall_performance_study",
    "request_for_information",
]

LANGS = ["de","en","fr","it"]

# Performans / nezaket
PAGE_SLEEP   = 0.20
DETAIL_SLEEP = 0.40
MAX_PAGES_PER_QUERY = 0  # 0 = sınırsız

# PDF ayarları
MAX_DESC_CHARS = 700
PDF_FONT_SIZE  = 8
PDF_LEADING    = 10
PAGE_SIZE      = landscape(A4)
MARGINS        = dict(leftMargin=24, rightMargin=24, topMargin=24, bottomMargin=24)
COLUMN_WIDTHS  = [120, 120, 80, 220, 300, 140, 100, 180]

def pick(d: Dict, langs: Iterable[str] = ("de","en","fr","it")) -> str:
    return next((d.get(k) for k in langs if isinstance(d, dict) and d.get(k)), "") if isinstance(d, dict) else ""

def clean_html(s: Optional[str]) -> str:
    s = "" if s is None else str(s)
    s = html.unescape(s)
    s = re.sub(r'</p\s*>', '\n', s, flags=re.I)
    s = re.sub(r'<\s*br\s*/?>', '\n', s, flags=re.I)
    s = re.sub(r'<[^>]+>', '', s)
    return s.strip()

def session_with_retries() -> requests.Session:
    s = requests.Session()
    retries = Retry(total=4, backoff_factor=0.4, status_forcelist=(429,500,502,503,504))
    s.mount("https://", HTTPAdapter(max_retries=retries))
    s.headers.update({"User-Agent": "SIMAPFetcher/1.4 (+https://example.org)"})
    return s

def build_params(last_item: Optional[str]=None, keyword: Optional[str]=None) -> Dict:
    """
    Search kullanmıyoruz; quick-filter için projectSubTypes veriyoruz.
    Böylece 'search en az 3 karakter' zorunluluğu olmadan arama yapılabiliyor.
    """
    params: Dict = {
        "projectSubTypes": PROJECT_SUBTYPES,
        "newestPublicationFrom":  NEWEST_PUBLICATION_FROM,
        "newestPublicationUntil": NEWEST_PUBLICATION_UNTIL,
        "lang": LANGS,
    }
    if CPV_CODES:
        params["cpvCodes"] = CPV_CODES
    if last_item:
        params["lastItem"] = last_item
    # İstersen ayrıca keyword filtresi de uygula (opsiyonel)
    if keyword and len(keyword.strip()) >= 3:
        params["search"] = keyword.strip()
    return params

def search_all_pages(sess: requests.Session, keyword: Optional[str]=None) -> List[Dict]:
    all_items: List[Dict] = []
    last_item: Optional[str] = None
    page = 0
    while True:
        if MAX_PAGES_PER_QUERY and page >= MAX_PAGES_PER_QUERY:
            break
        page += 1
        r = sess.get(SEARCH_ENDPOINT, params=build_params(last_item, keyword), timeout=30)
        r.raise_for_status()
        data = r.json() or {}
        batch = data.get("projects", []) or []
        if not batch:
            break
        all_items.extend(batch)
        pag = data.get("pagination") or {}
        next_last_item = pag.get("lastItem")
        print(f"[info] page {page} ({keyword or 'no-keyword'}): +{len(batch)} items, next={next_last_item}")
        if not next_last_item or next_last_item == last_item:
            break
        last_item = next_last_item
        time.sleep(PAGE_SLEEP)
    return all_items

def fetch_detail(sess: requests.Session, pid: str, pub: str) -> Dict:
    url = DETAIL_ENDPOINT.format(pid=pid, pub=pub)
    r = sess.get(url, timeout=30)
    r.raise_for_status()
    return r.json() or {}

def extract_fields(detail: Dict, list_level_title: str = "", fallback_pub_date: str = "") -> Dict[str, str]:
    pr = detail.get("procurement", {}) or {}
    title = pick(pr.get("projectTitle")) or pick(detail.get("title")) or list_level_title or ""
    desc_full = clean_html(pick(pr.get("orderDescription")))
    desc_pdf  = (desc_full[:MAX_DESC_CHARS].rstrip() + " … [truncated]") if len(desc_full) > MAX_DESC_CHARS else desc_full
    cpv_label = pick((pr.get("cpvCode") or {}).get("label"), ("en","de","fr","it"))
    process   = pr.get("processType", "") or ""
    pub_date  = detail.get("publicationDate") or fallback_pub_date or ""
    office    = pick(pr.get("procOfficeName"), ("de","en","fr","it")) or pick(detail.get("procOfficeName"))
    return dict(title=title, desc_full=desc_full, desc_pdf=desc_pdf,
                cpv_label=cpv_label or "", process=process, pub_date=pub_date, office=office or "")

def write_csv(records: List[Dict[str, str]], path: str) -> None:
    header = ["Project ID","Publication ID","Process Type","Title","Order Description",
              "CPV Label","Publication Date","Procurement Office"]
    with open(path, "w", newline="", encoding="utf-8") as f:
        w = csv.writer(f); w.writerow(header)
        for r in records:
            w.writerow([r["pid"], r["pub"], r["process"], r["title"],
                        r["desc_full"], r["cpv_label"], r["pub_date"], r["office"]])

def write_pdf(records: List[Dict[str, str]], path: str) -> None:
    styles = getSampleStyleSheet()
    tiny = ParagraphStyle("Tiny", parent=styles["Normal"],
                          fontSize=PDF_FONT_SIZE, leading=PDF_LEADING, wordWrap="CJK")
    def cell(x: str) -> Paragraph: return Paragraph(escape(x or ""), tiny)
    header = ["Project ID","Publication ID","Process Type","Title","Order Description",
              "CPV Label","Publication Date","Procurement Office"]
    table_data = [header] + [[cell(r[k]) for k in
                     ("pid","pub","process","title","desc_pdf","cpv_label","pub_date","office")]
                    for r in records]
    doc = SimpleDocTemplate(path, pagesize=PAGE_SIZE, title="SIMAP Tenders", **MARGINS)
    tbl = Table(table_data, repeatRows=1, colWidths=COLUMN_WIDTHS)
    tbl.setStyle(TableStyle([
        ("BACKGROUND",(0,0),(-1,0),colors.black),
        ("TEXTCOLOR",(0,0),(-1,0),colors.whitesmoke),
        ("FONTNAME",(0,0),(-1,0),"Helvetica-Bold"),
        ("FONTSIZE",(0,0),(-1,0),9),
        ("ALIGN",(0,0),(-1,0),"CENTER"),
        ("VALIGN",(0,0),(-1,-1),"TOP"),
        ("GRID",(0,0),(-1,-1),0.4,colors.grey),
        ("LEFTPADDING",(0,0),(-1,-1),5),
        ("RIGHTPADDING",(0,0),(-1,-1),5),
        ("TOPPADDING",(0,0),(-1,-1),2),
        ("BOTTOMPADDING",(0,0),(-1,-1),3),
    ]))
    doc.build([tbl])

def main():
    sess = session_with_retries()
    print(f"[info] Tarih filtresi: {NEWEST_PUBLICATION_FROM} .. {NEWEST_PUBLICATION_UNTIL}")

    dedup: "OrderedDict[Tuple[str,str], Dict]" = OrderedDict()
    title_cache: Dict[str, str] = {}

    # 1) Keyword’lü aramalar (opsiyonel)
    for kw in KEYWORDS:
        kw = kw.strip()
        if len(kw) < 3: continue
        hits = search_all_pages(sess, keyword=kw)
        for p in hits:
            pid, pub = p.get("id"), p.get("publicationId")
            if not pid or not pub: continue
            if p.get("title"): title_cache[pid] = pick(p.get("title"))
            dedup[(pid, pub)] = p
        print(f"[info] keyword '{kw}': uniq={len(dedup)}")

    # 2) Keyword’süz geniş tarama (sadece quick-filter’larla)
    wide_hits = search_all_pages(sess, keyword=None)
    for p in wide_hits:
        pid, pub = p.get("id"), p.get("publicationId")
        if not pid or not pub: continue
        if p.get("title"): title_cache[pid] = pick(p.get("title"))
        dedup[(pid, pub)] = p
    print(f"[info] wide-scan: +{len(wide_hits)} → uniq={len(dedup)}")

    if not dedup:
        print("[info] Kayıt bulunamadı. PROJECT_SUBTYPES’ı genişletin veya CPV_CODES=[] olduğuna emin olun.")
        return

    # 3) Detay ve kayıtlar
    records: List[Dict[str, str]] = []
    for j, ((pid, pub), p) in enumerate(dedup.items(), 1):
        try:
            detail = fetch_detail(sess, pid, pub)
        except Exception as e:
            print(f"[warn] detail {pid}/{pub}: {e}", file=sys.stderr)
            detail = {}
        list_title    = title_cache.get(pid, "")
        list_pub_date = p.get("publicationDate") or ""
        fields = extract_fields(detail, list_title, fallback_pub_date=list_pub_date)
        records.append({"pid": pid or "", "pub": pub or "", **fields})
        if j % 5 == 0:
            time.sleep(DETAIL_SLEEP)

    write_csv(records, "tenders_2018.csv")
    write_pdf(records, "tenders_2018.pdf")
    print(f"[done] 2018 CSV+PDF hazır: {len(records)} satır")

if __name__ == "__main__":
    main()
```
